```python
import pandas as pd
```

```python
data = pd.read_csv('Heart.csv')
```

```python
# mean
# Note: The mean is bigger than the trimmed mean, which is bigger than the median.
data['chol'].mean()
```

```python
# trim mean
from scipy.stats import trim_mean
trim_mean(data['chol'], 0.2)
```

```python
# median
data['chol'].median()
```
